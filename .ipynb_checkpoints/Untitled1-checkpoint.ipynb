```python
import os
import glob

for path in glob.glob('results/esc50_gap_7/th*/results.txt'):
    with open(path, 'r') as f:
        l = f.readlines()
        if len(l) > 0:
            print(path)
            TP = float(l[2][2:l[2].find('\t')])
            FN = float(l[2][l[2].find('FN')+2 : l[2].find('\n')])
            FP = float(l[3][3:l[3].find('\t')])
            TN = float(l[3][l[3].find('TN')+2 : l[3].find('\n')])
            print('TP {:2.2f}% | FN {:2.2f}%'.format(TP*100, FN*100))
            print('FP {:2.2f}% | TN {:2.2f}%'.format(FP*100, TN*100))
            
            print('precision = ', TP/(TP+FP))
            print('recall = ', TP/(TP+FN))
            print('----')
```

```python
from argparse import Namespace
import numpy as np

def get_rand_args():
    args = Namespace(
        save='./results/esc50_gap_7',
        split=[1, 2, 3],
        noiseAugment=bool(np.random.randint(2)),
        inputLength=0,
        act_thrld=np.random.uniform(20, 40),
        act_window=np.random.randint(5, 10),
        min_act_per_window=np.random.randint(2, 5),
        use_zero_activations=bool(np.random.randint(2)),
        n_batch_eval=100,
        store_cm=False)
    if args.min_act_per_window > args.act_window:
        args.min_act_per_window = args.act_window

    args._str = 'th{}_win{}_actPwin{}_inLen{}_nBatch{}_nAug{}_zeroAct{}'.format(
            args.act_thrld, args.act_window, args.min_act_per_window,
            args.inputLength, args.n_batch_eval, args.noiseAugment, args.use_zero_activations)
    return args
```

```python
from shutil import copy
from guppy import hpy
import os
import pandas as pd
import test; reload(test)
import gc
h = hpy() 

results_name = 'ommission, insertion, acc, FN, TP, FP, TN, conf_matrix, precision, rappel'.split(',')
df_path = os.path.join('./results/exp.csv')

# Read csv if any
df = None
if os.path.isfile(df_path):
    df = pd.read_csv(df_path)

for _ in range(10):
    args = get_rand_args()
    results = test.main2(get_rand_args())
    params = vars(args)
    params['split'] = str(params['split'])
    params.update(results)
    
    # Update csv
    if df is None:
        params['split'] = [-1]
        df = pd.DataFrame.from_dict(params)
    else:
        df = df.append(params, ignore_index=True)
    
    # Save csv
    print('-- saving --')
    if os.path.isfile(df_path):
        copy(df_path, df_path + '_old')
    with open(df_path, 'w') as f:
        df.to_csv(f)
        
    print h.heap()
    gc.collect()


```
