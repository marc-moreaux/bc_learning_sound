```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from itertools import compress


def str2arr(mstr):
    if type(mstr) is str and len(mstr) > 1000:
        arr = np.array(np.matrix(mstr))
        return arr
    return 0


def cleanDF(df):
    # remove unnamed columns
    cols_keep = [c for c in df.columns if 'Unnamed' not in c]
    df = df[cols_keep]
    
    # where are the conf matrix...
    has_cm = df['1-cm'].str.len() > 1000
    has_cm = has_cm
    return df, has_cm


# Load csv and drop undesired columns
df = pd.read_csv('./results/exp.csv')
df, has_cm = cleanDF(df)
```

```python
# generate new columns
df['TP'] = df.filter(regex=(".-TP")).mean(axis=1)
df['TN'] = df.filter(regex=(".-TN")).mean(axis=1)
df['FP'] = df.filter(regex=(".-FP")).mean(axis=1)
df['FN'] = df.filter(regex=(".-FN")).mean(axis=1)
df['TPR'] = df.TP / (df.TP + df.FN)
df['FPR'] = df.FP / (df.FP + df.TN)
df['TNR'] = df.TN / (df.TN + df.FP)
df['FNR'] = df.FN / (df.FN + df.TP)

df['precision'] = df.TP / (df.TP + df.FP)
df['rappel'] = df.TP / (df.TP + df.FN)
df['F1'] = 2 * ((df.precision * df.rappel) / (df.precision + df.rappel))
df['min_act_per_window'] = map(int, df['min_act_per_window'])
cms = [df[k].apply(str2arr) for k in df.filter(regex=(".-cm"))]
cms = map(sum,zip(*cms))
```

```python
res = df[(df.act_window > 0) &
         (df.min_act_per_window > 0) &
         (df.noiseAugment == False) &
         (df.use_zero_activations == False) & 
         has_cm]

plt.scatter(res.act_thrld, res.F1)
plt.scatter(res.act_thrld, res.TP)
plt.scatter(res.act_thrld, res.FP)
plt.legend()
plt.xlabel('activation threshold')
plt.ylim(0, 1)
plt.show()

plt.scatter(res.FPR, res.TPR)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.show()


selection = [u'act_thrld', u'act_window', u'inputLength', u'min_act_per_window',
             u'noiseAugment', u'use_zero_activations', u'precision', u'rappel', 
             u'F1', u'TP', u'TN', u'FP', u'FN', u'TPR', u'FPR', u'TNR', u'FNR']

res[selection]
```

```python
def cm2ratio(cm, class_idx):
    idxs = np.arange(len(cm))
    cm = np.array(cm)
    cm = cm / cm.sum()
    ratio = {'tp' : cm[class_idx        ][   class_idx        ],
             'tn' : cm[idxs != class_idx][:, idxs != class_idx].sum(),
             'fp' : cm[idxs != class_idx][:, class_idx        ].sum(),
             'fn' : cm[class_idx        ][  idxs != class_idx].sum()}
    ratio['TPR'] = ratio['tp'] / (ratio['tp'] + ratio['fn']) * 100
    ratio['FPR'] = ratio['fp'] / (ratio['fp'] + ratio['tn']) * 100
    return ratio


res = df[(df.act_window > 0) &
         (df.min_act_per_window > 0) &
         (df.noiseAugment == False) &
         (df.use_zero_activations == False) & 
         has_cm]


_cms = [cms[idx] for idx in res.index]
ratios = [cm2ratio(cm, cIdx) for cm in _cms for cIdx in range(1,2)]

print ratios[0]

rat = pd.DataFrame(ratios)
plt.scatter(rat.FPR, rat.TPR)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.xlim(0, 10)
plt.ylim(0, 100)
plt.show()
```

```python
import cPickle

with open('./results/exp.pkl', 'rb') as f:
    mdicts = cPickle.load(f)
print(mdicts[0].keys())
```

```python
from sklearn.metrics import roc_curve, auc

fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(-1, 50):
    fpr[i] = [0, 1]
    tpr[i] = [0, 1]

for mdict in mdicts:
    y = mdict['1-gt'].ravel()
    pred = mdict['1-pred'].ravel()

    for i in range(-1, 50):
        roc = roc_curve(y == i, pred == i)
        fpr[i].append(roc[0][1])
        tpr[i].append(roc[1][1])

for i in range(-1, 50):
    fpr[i].sort()
    tpr[i].sort() #THIS IS A FAIL
    roc_auc[i] = auc(fpr[i], tpr[i])
```

```python
plt.figure()
lw = 2
plt.plot(fpr[2], tpr[2], color='darkorange',
         lw=lw, label='ROC curve (area = %0.2f)' % roc_auc[1])
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()
```

```python
for i in range(-1, 50):
    plt.plot(fpr[i], tpr[i],
             label='ROC curve of class {0} (area = {1:0.2f})'
             ''.format(i, roc_auc[i]))

plt.plot([0, 1], [0, 1], 'k--', lw=lw)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Some extension of Receiver operating characteristic to multi-class')
#plt.legend(loc="lower right")
plt.show()
```
